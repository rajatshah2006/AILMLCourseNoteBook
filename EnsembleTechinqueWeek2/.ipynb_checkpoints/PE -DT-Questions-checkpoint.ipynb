# 6. See the distribution of 'Class' variable and plot it using appropriate graph

```python
def create_labeled_count_plot(data, x_col, hue_col, title='Labeled Count Plot', xlabel='Category', ylabel='Count', palette='viridis'):
    """
    Create a labeled count plot with hue using Seaborn.

    Parameters:
    - data: DataFrame containing the data
    - x_col: str, name of the column to be used for x-axis
    - hue_col: str, name of the column to be used for hue
    - title: str, title of the plot
    - xlabel: str, label for the x-axis
    - ylabel: str, label for the y-axis
    - palette: str, color palette for the bars

    Returns:
    - None
    """
    plt.figure(figsize=(10, 6))
    count_plot = sns.countplot(x=x_col, hue=hue_col, data=data, palette=palette)

    # Adding labels to the bars
    for p in count_plot.patches:
        height = p.get_height()
        count_plot.annotate(format(height, '.0f'), 
                            (p.get_x() + p.get_width() / 2., height), 
                            ha='center', va='center', 
                            xytext=(0, 10), 
                            textcoords='offset points')

    # Adding title and labels
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    # Show plot
    plt.show()
```

```python
create_labeled_count_plot(df1,x_col='class',hue_col='class')
```

# 8. Split the pima_df into training and test set in the ratio of 70:30 (Training:Test).

```python
X=df1.drop('class',axis=1)
y=df1.pop('class')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.30, random_state=1,stratify=y)
```

# 9. Create the decision tree model using entropy method of reducing the entropy and fit it to training data.

```python
dTree = DecisionTreeClassifier(criterion = 'entropy', random_state=1,max_depth=3)
dTree.fit(X_train, y_train)
```

# 10. Print the accuracy of the model & print the confusion matrix

```python
df1.columns
```

```python
print(dTree.score(X_train, y_train))
print(dTree.score(X_test, y_test))
```

```python
from sklearn.tree import  plot_tree
plt.figure(figsize=(20, 10))
plot_tree(dTree, feature_names=X.columns, class_names=['0','1'], filled=True)
plt.title('Decision Tree Visualization')
plt.show()
```

```python
from sklearn import metrics
y_predict = dTree.predict(X_test)
cm=metrics.confusion_matrix(y_test, y_predict, labels=[0, 1])

df_cm = pd.DataFrame(cm, index = [i for i in ["0","1"]],
                  columns = [i for i in ["0","1"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True ,fmt='g')
```

# 11. Apply the Random forest model and print the accuracy of Random forest Model

```python
from sklearn.ensemble import RandomForestClassifier
rfcl = RandomForestClassifier(n_estimators = 50, random_state=1)
rfcl = rfcl.fit(X_train, y_train)
```

```python
y_predict = rfcl.predict(X_test)
print(rfcl.score(X_test, y_test))
```

```python
cm=metrics.confusion_matrix(y_test, y_predict,labels=[0, 1])

df_cm = pd.DataFrame(cm, index = [i for i in ["0","1"]],
                  columns = [i for i in ["0","1"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True ,fmt='g')
```
