```python
df_merge=pd.merge(movie,rating,on='movie id',how='inner')
```

```python
df_merge.head()
```

```python
movie_mean_fram=pd.DataFrame()
```

```python
movie_mean_fram.head()
```

```python
for i in Genere:
    movie_mean_fram[i]=np.where(df_merge[i]==1,df_merge['rating'],np.nan)

movie_mean_fram.head()
```

```python
movie_mean_fram.mean(axis=0).sort_values(ascending=False)
```

```python
avg_rating = pd.DataFrame(df_merge.groupby(['movie title'])['rating'].mean().reset_index())
avg_rating
```

```python
avg_rating['Avg rating']=avg_rating['rating']
```

```python
avg_rating
```

```python
avg_rating.drop('rating',axis=1,inplace=True)
```
