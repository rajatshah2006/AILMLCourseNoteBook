```python
def hist_box_plot(data,xaxis,kde=False,catgorial=False):
    plt.xticks(rotation=90)
    sns.histplot(data=data,x=xaxis,kde=kde)
    plt.show()
    if catgorial==True:
        plt.xticks(rotation=90)
        sns.countplot(data=data,x=xaxis)
        plt.show()        
    else:    
        sns.boxplot(data=data,x=xaxis)
        plt.show()
        sns.scatterplot(data=data,x=xaxis)
        plt.show()
```

```python
hist_box_plot(df,df['age'],True)
```

Age plot is rightly skewed telling that 
  Median age is around 38 
  Q3+1.5IQR is aroung 79 ..
  Young Population: The peak of the histogram, representing the mode or most common age group, is towards the lower end of the age range. This suggests that the population consists mostly of younger individuals.
  
   Fewer Older Individuals: As the histogram extends towards higher ages, the frequency of individuals decreases. There are relatively fewer individuals in older age groups compared to younger age groups.
