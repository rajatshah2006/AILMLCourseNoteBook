```python
sns.histplot(data=df,x='price')
```

## Note
Histograms can be used to examine the distribution of numeric variables. They help in understanding the amount of data present in specific numeric intervals for a variable. Since Education level is a categorical variable, a histogram is not appropriate for that.

```python
plt.title("Histogram of Price of Car")
plt.xlim(0,50000)
plt.ylim(0,70)
plt.xlabel('price of car')
plt.ylabel('Number of Cars in price range')
sns.histplot(df,x='price',color="orange");
```

# to change the bin size 
## Need to pass bins option in histplot

```python
sns.histplot(data=df,x='price',bins=20)
```

# binwidth
if we want to speciy width of bar that can be done using binwidth option

## Right and left skewed
If mean is closer to Q1 and left whisker is smaller then right whisker to minimum then graph is right skewed
If mean is closer to Q2 and right whisker is smaller then left whisker to minimum then graph is right skewed


```python
sns.boxplot(df,x='price',y="body_style")
```
